```python
import cv2
import matplotlib.pyplot as plt

def processingImage(image_path, show=True):
    # Read image
    raw_image = cv2.imread(image_path)

    # Safety check
    if raw_image is None:
        raise FileNotFoundError(f"Image not found or unreadable: {image_path}")
    raw_image_rgb = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
    
    image_bliur = cv2.GaussianBlur(src=raw_image_rgb,ksize=(5,5),sigmaX=2,sigmaY=2)
    image_bliur_rgb = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)

    
    

    # Display
    if show:
        plt.figure(figsize=(6, 6))
        plt.imshow(image_bliur_rgb)
        plt.axis("off")
        plt.title("Original Image")
        plt.show()

    return raw_image, raw_image_rgb
```
